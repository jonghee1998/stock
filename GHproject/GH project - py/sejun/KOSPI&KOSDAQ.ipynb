```python
## KOSPI 불러오기

import yfinance as yf
import datetime

today = datetime.date.today()

# Define the ticker symbol
ticker_symbol = "^KS11"

# Get data on the specified ticker
ticker_data = yf.Ticker(ticker_symbol)

# Get the historical prices for this ticker
ticker_df = ticker_data.history(period='1d', start='2018-1-1', end=today)
ticker_df.index = ticker_df.index.tz_localize(None).normalize()
ticker_df = ticker_df.sort_index(ascending=False)

print(ticker_df)
ticker_df.to_csv('C:\\Users\\yss06\\Desktop\\python\\loaded files\\KOSPI 지수.csv', encoding="utf-8-sig")

```

```python
import yfinance as yf
import datetime

today = datetime.date.today()

# KOSDAQ 데이터 가져오기
ticker_symbol_kosdaq = "^KQ11"
ticker_data_kosdaq = yf.Ticker(ticker_symbol_kosdaq)
ticker_df_kosdaq = ticker_data_kosdaq.history(period='1d', start='2018-1-1', end=today)
ticker_df_kosdaq.index = ticker_df_kosdaq.index.tz_localize(None).normalize()
ticker_df_kosdaq = ticker_df_kosdaq.sort_index(ascending=False)

# KOSPI 데이터 가져오기
ticker_symbol_kospi = "^KS11"
ticker_data_kospi = yf.Ticker(ticker_symbol_kospi)
ticker_df_kospi = ticker_data_kospi.history(period='1d', start='2018-1-1', end=today)
ticker_df_kospi.index = ticker_df_kospi.index.tz_localize(None).normalize()
ticker_df_kospi = ticker_df_kospi.sort_index(ascending=False)

# 각 데이터프레임의 컬럼명에 접두어 추가
ticker_df_kosdaq.columns = ["KOSDAQ_" + col for col in ticker_df_kosdaq.columns]
ticker_df_kospi.columns = ["KOSPI_" + col for col in ticker_df_kospi.columns]

# 두 데이터프레임 합치기
master_df = ticker_df_kospi.join(ticker_df_kosdaq, how="outer")
master_df = master_df.sort_index(ascending=False)

print(master_df)
master_df.to_csv('C:\\Users\\yss06\\Desktop\\python\\loaded files\\KOSPI+KOSDAQ.csv', encoding="utf-8-sig")
```
